```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
```

## Generating Data

```python
#X,Y = make_blobs(n_features=2, n_samples=500, centers=2, random_state=1)

X,Y = make_moons(n_samples=1000, shuffle=True, noise=0.2, random_state=1)

print(X.shape, Y.shape)
```

```python
plt.style.use('seaborn')
plt.scatter(X[:,0], X[:,1], c=Y, cmap=plt.cm.Accent)
plt.show()
```

## Perceptron Model

```python
def sigmoid(z):
    return (1.0)/(1 + np.exp(-z))
```

```python
## Broadcasting - In numpy arrays
z = np.array([1,2,3,4,5])
sigmoid(z)
```

```python
def predict(X, weights):
    z = np.dot(X, weights)
    predictions = sigmoid(z)
    return predictions

def loss(X,Y,weights):
    """Binary Cross Entropy"""
    Y_ = predict(X,weights)
    cost = np.mean(-Y * np.log(Y_) - (1-Y) * np.log(1-Y_))
    return cost

def update(X,Y,weights,learning_rate):
    """Perform weight updates for 1 epoch"""
    Y_ = predict(X,weights)
    dw = np.dot(X.T, Y_ - Y)
    
    m = X.shape[0]
    weights = weights - learning_rate*dw/(float(m))
    
    return weights

def train(X,Y,learning_rate=0.5,maxEpochs=100):
    
    #Modify the input to handle the bias term
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    #Initialize weights
    weights = np.zeros(X.shape[1])
    
    for epoch in range(maxEpochs):
        weights = update(X,Y,weights,learning_rate)
        
        if epoch%10==0:
            l = loss(X,Y,weights)
            print("Epoch %d Loss %.4f"%(epoch,l))
            
    return weights
```

```python
weights = train(X,Y,learning_rate=0.8,maxEpochs=1000)
```

```python
def getPredictions(X_Test,weights,labels=True):
    if X_Test.shape[1] != weights.shape[0]:
        ones = np.ones((X_Test.shape[0],1))
        X_Test = np.hstack((ones,X_Test))
        
    probs = predict(X_Test, weights)
    
    if not labels:
        return probs
    else:
        labels = np.zeros(probs.shape)
        labels[probs>=.5] = 1
        return labels
```

```python
x1 = np.linspace(-2,2,10)
x2 = -(weights[0] + weights[1]*x1)/weights[2]
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap=plt.cm.Accent)
plt.plot(x1,x2,c='red')
plt.show()
```

## Accuracy

```python
Y_ = getPredictions(X,weights)
training_acc = np.sum(Y_ == Y)/Y.shape[0]
print(training_acc)
```
