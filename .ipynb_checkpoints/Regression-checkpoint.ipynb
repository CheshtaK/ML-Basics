```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Data(X) - Acidity of Milk and Data(Y) - Density of Milk

```python
def readData(filename):
    df = pd.read_csv(filename)
    return df.values

x = readData('./data/linearX.csv')
y = readData('./data/linearY.csv')
x = x.reshape((99,))
y = y.reshape((99,))
print(x)
print(y)
```

```python
### Plotting the data
plt.scatter(x,y)
plt.show()
```

```python
##Normalization - Mean=0, Standard Deviation=1
x = x - x.mean()/(x.std())
plt.scatter(x,y)
plt.show()

X = x
Y = y
```

### Linear Regression Algorithm

```python
def hypothesis(theta, x):
    return theta[0] + theta[1]*x

def error(X,Y,theta):
    total_error = 0
    m = X.shape[0]
    
    for i in range(m):
        total_error += (Y[i] - hypothesis(theta, X[i]))**2
        
    return 0.5*total_error

def gradient(Y,X,theta):
    grad = np.array([0.0,0.0])
    m = X.shape[0]
    for i in range(m):
        grad[0] += -1*(Y[i] - hypothesis(theta, X[i]))
        grad[1] += -1*(Y[i] - hypothesis(theta, X[i]))*X[i]
        
    return grad

def gradientDescent(X,Y,learning_rate,max_iterations):
    grad = np.array([0.0,0.0])
    theta = np.array([0.0,0.0])
    
    e = []
    
    for i in range(max_iterations):
        grad = gradient(Y,X,theta)
        
        ce = error(X,Y,theta)
        
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        
        e.append(ce)
        
    return theta,e

theta,e = gradientDescent(X,Y,learning_rate=0.001, max_iterations=100)
print(theta[0], theta[1])
```

```python
plt.scatter(X,Y)
plt.plot(X,hypothesis(theta,X),color='r')
plt.show()
```
