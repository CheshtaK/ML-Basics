```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dfx = pd.read_csv('./data/linearX.csv')
dfy = pd.read_csv('./data/linearY.csv')

x = dfx.values
y = dfy.values

#Normalising
x2 = (x-x.mean())/x.std()

plt.scatter(x,y,label="Raw Data")
plt.scatter(x2,y,color='orange',label="Normalised Data"
           )
plt.legend()
plt.show()
```
