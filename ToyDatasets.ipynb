```python
from sklearn.datasets import load_boston, make_classification, make_regression, make_blobs
```

```python
boston = load_boston()
```

```python
XB = boston.data
YB = boston.target
```

```python
print(XB.shape)
print(YB.shape)
```

```python
plt.figure(0, figsize=(8,8))

plt.subplot(221)
X, Y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_classes=2)
plt.scatter(X[:,0], X[:,1], c=Y, edgeColor='k')

plt.subplot(222)
X1, Y1 = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_classes=3, n_clusters_per_class=1)
plt.scatter(X1[:,0], X1[:,1], c=Y1, edgeColor='k')

plt.subplot(223)
X2, Y2 = make_blobs(n_features=2, centers=3)
plt.title("Blobs")
plt.scatter(X2[:,0], X2[:,1], c=Y2, edgeColor='k')

plt.subplot(224)
X3, Y3 = make_regression(n_features=1, n_samples=200, bias=4, noise=15)
plt.title("Regression")
plt.scatter(X3, Y3, color='orange', edgeColor='k')


plt.show()
```
