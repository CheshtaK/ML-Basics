## Generating Classification Data

```python
u1 = np.array([2,4])
cov_1 = np.array([[2,-0.8],
                 [-0.8,1]])

u2 = np.array([-1,-3])
cov_2 = np.array([[1,0],
                 [0,2]])
```

```python
X1 = np.random.multivariate_normal(u1,cov_1,300)

X2 = np.random.multivariate_normal(u2,cov_2,400)
```

```python
plt.scatter(X1[:,0], X1[:,1],label="Apple")
plt.scatter(X2[:,0], X2[:,1],marker='^',label="Orange")
plt.legend()
plt.show()
```

```python
y = np.zeros(700)
y[:300] = 1
```

```python
X = np.vstack((X1,X2))
```

```python
#plt.scatter(X[:,0],X[:,1])

for i in range(700):
    if y[i] == 1:
        plt.scatter(X[i,0], X[i,1], color='green')
    else:
        plt.scatter(X[i,0], X[i,1], color='red')


plt.show()
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

## Generating Data using Sk-Learn

```python
from sklearn.datasets import make_regression, make_classification, make_blobs, make_moons
```

### Regression

```python
X,Y = make_regression(n_samples=100, n_features=1, n_informative=1, noise=2.8)
print(X.shape, Y.shape)
```

```python
plt.scatter(X,Y)
plt.show()
```

### Classification

```python
XC, YC = make_classification(n_samples=400, n_classes=3, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1,random_state=5)
```

```python
plt.scatter(XC[:,0], XC[:,1], c=YC)
plt.show()
```

### Blobs Dataset

```python
XB, YB = make_blobs(n_samples=1000, n_features=2, centers=5, random_state=10)
```

```python
plt.scatter(XB[:,0], XB[:,1], c=YB)
plt.show()
```

### Moons Dataset

```python
XM, YM = make_moons(n_samples=100, noise=0.1, random_state=5)
```

```python
plt.scatter(XM[:,0], XM[:,1], c=YM)
plt.show()
```
